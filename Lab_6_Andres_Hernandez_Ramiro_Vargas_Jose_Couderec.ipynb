```python

# Visualización de la evolución de la edad media en Guatemala
 

import pandas as pd
import plotly.express as px


df = pd.read_csv("edadmedia.csv")


df.head()

#Datos en Formato Largo
df_long = df.melt(id_vars=["Year"], var_name="Departamento", value_name="EdadMedia")

# Gráfica
fig = px.line(
    df_long,
    x="Year",
    y="EdadMedia",
    color="Departamento",
    title="Evolución de la Edad Media por Departamento en Guatemala",
    labels={"Year": "Año", "EdadMedia": "Edad media (años)"},
    markers=True
)

# Ponerla Bonita
fig.update_layout(
    title_font_size=20,
    legend_title_text="Departamento",
    plot_bgcolor="white",
    hovermode="x unified"
)

# Mostrar la gráfica
fig.show()
```
